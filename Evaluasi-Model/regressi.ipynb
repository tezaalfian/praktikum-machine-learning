```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('../dataset/data-penjualan.csv',';')
dataset
```

```python
x = np.array(dataset.promosi)
y = np.array(dataset.penjualan)

x_train = x.reshape(-1,1)
y_train = y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_train, y_train, color = 'blue')
plt.plot(x_train, model.predict(x_train), color = 'red')
plt.xlabel('Promosi')
plt.ylabel('Penjualan')
plt.title("Perbandingan Promosi dan Penjualan")
plt.show()
```

```python
x_test = np.array([62,73,82,90,100]).reshape(-1,1)
y_test = np.array([160, 170, 180, 190, 200])
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)

print("MSE : ", mse)
print("RMSE : ", rmse)
print("MAE : ", mae)
```
